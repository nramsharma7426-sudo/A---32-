```python
#Q.8
n = 1221
tn = n
revn = 0
while(tn>0):
    revn= revn*10+tn%10
    tn = tn//10
if revn == n:
    print("palindrome")
else:
    print("not palindrome")
```

```python
#Q. 13 ***
string = "Hi 25"
upper = lower = digit = space = 0
for x in string:
    if x.isupper():
        upper+=1
    elif x.islower():
        lower+=1
    elif "0"<=x<="9":
        digit+=1
    else:
        space+=1
print(upper)
print(lower)
print(digit)
print(space)
```
